# Validating the 'Year' column

```python
import matplotlib.pyplot as plt
avengers['Year'].hist()
```

I can tell that some data was entered in error. Marvel comics started in the 1960s, so no characters could have been introduced before then. In the cell below, I subset the dataframe to include characters introduced in 1960 or later.

```python
true_avengers = avengers[avengers['Year'] > 1959]
```

```python
def clean_deaths(row):
    num_deaths = 0
    columns = ['Death1', 'Death2', 'Death3', 'Death4', 'Death5']
    
    for c in columns:
        death = row[c]
        if pd.isnull(death) or death == 'NO':
            continue
        elif death == 'YES':
            num_deaths += 1
    return num_deaths

true_avengers['Deaths'] = true_avengers.apply(clean_deaths, axis=1)
```

Next, I'll check that the 'Years since joining' value was accurately calculated. We've got 159 rows in true_avengers. Do any of those have an inaccurate value for 'Years since joining'?

```python
true_avengers['years_since_joining_2'] = 2015 - true_avengers['Year']
true_avengers['bool'] = true_avengers['Years since joining'] == true_avengers['years_since_joining_2']
wrong_years_since_joining = true_avengers[true_avengers['bool'] == False]
wrong_years_since_joining.shape
```

Printing the shape of the resulting dataframe shows there are no rows where 'Years since joining' was inaccurate. Below, I demonstrate an alternate way to reach the same conclusion:

```python
joined_accuracy_count  = 0
for index, row in true_avengers.iterrows():
    correct_joined_years = 2015 - row['Year']
    if correct_joined_years == row['Years since joining']:
        joined_accuracy_count += 1
        
print(joined_accuracy_count)
```

I know all rows had an accurate 'Years since joining value' because joined_accuracy_count incremented from 0 to 159
